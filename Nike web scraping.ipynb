# Import the required libraries

```python
#importing required libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# Send request to the website server 

```python
page = requests.get('https://www.nike.com/gb/w/black-friday-shoes-2phlmzy7ok')
page
```

# Page contents

```python
soup = BeautifulSoup(page.content)
soup
```

```python
#Scrapping the shoe
name = soup.find('a', class_="product-card__link-overlay")
name
```

```python
name.text
```

```python
#Scrapping the price
price = soup.find('div', class_="product-card__animation_wrapper")
price
```

```python
price.text
```

```python
#Scrapping the number of available color
color = soup.find('div', class_="product-card__count-wrapper false false")
color
```

```python
color.text
```

```python
#retain only the numeric value and make it an integer
no_of_color = int(''.join(filter(str.isdigit, color.text)))
no_of_color
```

```python
#Scrapping the category
cat = soup.find('div', class_="product-card__subtitle")
cat
```

```python
cat.text
```

# Scraping multiple 'names', 'price', 'no of color' and 'category'

```python
#scraping multiple names
names = []

for i in soup.find_all('a', class_="product-card__link-overlay"):
    names.append(i.text)

names
```

```python
#scraping multiple prices
prices = []

for i in soup.find_all('div', class_="product-card__animation_wrapper"):
    prices.append(i.text)

prices
```

```python
#scraping multiple number of available color
no_of_colors = []

for i in soup.find_all('div', class_="product-card__count-wrapper false false"):
    no_of_colors.append(i.text)

no_of_colors
```

```python
#scraping multiple category
category = []

for i in soup.find_all('div', class_="product-card__subtitle"):
    category.append(i.text)
    
category
```

```python
#Finding the length
print(len(names),len(prices),len(no_of_colors),len(category))
```

```python
#Making dataFrame
df = pd.DataFrame({'Name':names, 'Price':prices, 'Color': no_of_colors, 'Category':category})
df
```

```python
# To extract numeric part of Color column and convert to int
def extract_numeric(s):
    return int(''.join(filter(str.isdigit, s)))
```

```python
# Applying the function to the 'color' column
df['Color'] = df['Color'].apply(extract_numeric)
print(df)
```

```python
#To extract the first price

def extract_first_price(s):
    return s.split('£')[1]  # Split by '£' and keep the second part

# Applying the function to the 'price' column
df['Price'] = df['Price'].apply(extract_first_price)

print(df)
```
